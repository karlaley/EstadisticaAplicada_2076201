```python
# Librerias especializadas

from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
```

```python
# Importar los datos
datos1 = pd.read_excel('ejercicio_km.xlsx')
```

```python
## Estimar modelo Kaplan Meier
# En los datos siempre usamos el orden de 1. tiempo y 2. datos
km1 = KaplanMeierFitter().fit(datos1['tiempo_sup'], datos1['evento'])
```
