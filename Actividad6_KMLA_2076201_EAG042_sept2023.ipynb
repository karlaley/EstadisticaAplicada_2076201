```python
# Convertir la variable salario a dummy 

salarios_dummy = desercion['salario']
#Hacer dummies
salarios_dummy= pd.get_dummies(salarios_dummy, prefix='salario', drop_first = True).astype(int)
salarios_dummy.head()
```

```python
desercion = pd.concat([desercion, departamentos_dummy], axis=1)
desercion = pd.concat([desercion, salarios_dummy], axis=1)
```

**ESTIMACIÓN DE AFT ADECUADO**

```python
# verificar si es Weibull , exponencial o Lognormal. 

modelo_expo = ExponentialFitter().fit(desercion['semestre'],desercion['renunciar'])
modelo_weibull = WeibullFitter().fit(desercion['semestre'],desercion['renunciar'])
modelo_logn = LogNormalFitter().fit(desercion['semestre'],desercion['renunciar'])
```

```python
## Calcular los AIC de cada modelo

modelo_expo.AIC_, modelo_weibull.AIC_, modelo_logn.AIC_
```
