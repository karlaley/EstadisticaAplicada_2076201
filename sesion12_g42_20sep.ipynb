```python
## Librerias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Datos
## el parse_dates sirve para que en la columna que encuentre lo reconozca como fecha

pasajeros = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/airline_passengers.csv', parse_dates = True)
```

```python
## renombrar columnas
pasajeros = pasajeros.rename(columns = {'Month':'Mes', 'Thousands of Passengers':'Miles de pasajeros'})
```

```python
pasajeros
```

```python
plt.plot(pasajeros['Mes'], pasajeros['Miles de pasajeros'])
plt.title('Miles de pasajeros de avion en Estados Unidos de 1940-1960')
plt.show();
```
