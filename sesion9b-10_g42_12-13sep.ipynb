```python
# Librerias básicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Librerias especializadas
from lifelines import CoxPHFitter
import seaborn as sns
```

```python
# Importar datos
df_cancer = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/cancer.csv')
```

```python
df_cancer.head()
```

Los datos constan de 228 observaciones y 10 variables/columnas. La descripción de las variables es la siguiente:
inst: código de institución

**time (d1)**: tiempo de supervivencia en días

**status (d2)**: estado de censura 1 = censurado, 2 = muerto

**age (i1)**: Edad en años

**sex (i2)**: Masculino = 1 Femenino = 2

**ph.ecog (i3)**: puntuación de rendimiento ECOG según la calificación del médico. 0 = asintomático, 1 = sintomático pero completamente ambulatorio, 2 = en cama <50% del día, 3 = en cama > 50% del día pero no encamado, 4 = encamado

**ph.karno (i4)**: puntuación de desempeño de Karnofsky (mala = 0; buena = 100) calificada por el médico

**pat.karno (i4)**: puntuación de rendimiento de Karnofsky según la calificación del paciente

**meal.cal (i5)**: Calorías consumidas en las comidas

**wt.loss (i6)**: Pérdida de peso en los últimos seis meses

```python
# Depurar y estadisticas descriptivas
# recodificar las variables dummy
df_cancer['status'] = df_cancer['status']-1
df_cancer['sex'] = df_cancer['sex']-1

# status: 0=censurado, 1=muerto
# sex: 0=hombre, 1=mujer
```

```python
# borrar la variable inst
df_cancer = df_cancer.drop('inst', axis=1)
```

```python
# verificar que todas las variables sean numericas
df_cancer.dtypes
# todas las variables son númericas (enteras, decimales)
```

```python
# contar los valores perdidos/faltantes (missing values)
df_cancer.isnull().sum()
```

Cuando hay valores pérdidos hay dos opciones:
1. Eliminar las observaciones con valores perdidos
2. Imputar datos (promedio)

```python
# asignar a los valores perdidos el valor promedio de la variable
df_cancer['ph.karno']=df_cancer['ph.karno'].fillna(df_cancer['ph.karno'].mean())
df_cancer['pat.karno']=df_cancer['pat.karno'].fillna(df_cancer['pat.karno'].mean())
df_cancer['meal.cal']=df_cancer['meal.cal'].fillna(df_cancer['meal.cal'].mean())
df_cancer['wt.loss']=df_cancer['wt.loss'].fillna(df_cancer['wt.loss'].mean())
df_cancer=df_cancer.dropna()
df_cancer['ph.ecog']=df_cancer['ph.ecog'].astype('int64')
```

```python
# Estadisticas descriptivas
df_cancer.describe()
```

El 39.47% son mujeres, el 72.36% de las personas fallecieron, la edad promedio es de 62,44, la edad más joven es 39 y la edad más frande es 82. La media de ph.karno es mayor que la media pat.karno que significa que los pacientes consideran su estado de salud mucho peor de lo que realmente es.

```python
# Histrograma del tiempo de supervivencia
plt.hist(df_cancer['time'], bins = 30);
```

```python
sns.histplot(df_cancer['time'], bins = 30, kde=True)
sns.kdeplot(df_cancer['time'])
plt.xlim(0,1022)
plt.xlabel('días')
plt.ylabel('Frecuencia')
plt.title('Histograma de la variable time');
```

```python
sns.histplot(df_cancer['age'], bins = 30, kde=True)
sns.kdeplot(df_cancer['age'])
plt.xlim(39,82)
plt.xlabel('días')
plt.ylabel('Frecuencia')
plt.title('Histograma de la edad');
```

```python
# estimar el modelo 
cph1 = CoxPHFitter().fit(df_cancer, 'time', 'status')
```

```python
cph1.print_summary()
```

**age**, p-value: 0.18 no rechazamos H0, age no tiene efecto en el riesgo de morir de cáncer

**sex**, p-value<alfa rechazamos Ho, ser mujer reduce el riesgo de morir de cancer en 44% respecto a los hombres

**ph.karno**, p-value: 0.10. No rechazamos Ho, ph.karno no tiene efecto sobre el riesgo de morir de cancer

**pat.karno**, p-value: 0.08. No rechazamos Ho, pat.karno no tiene efecto sobre el riesgo de morir de cancer

**meal.cal**, p-value: 0.08. No rechazamos Ho, meal.cal no tiene efecto sobre el riesgo de morir de cancer

**wt.loss**, p-value: 0.08. No rechazamos Ho, wt.loss no tiene efecto sobre el riesgo de morir de cancer

**ph.ecog** (suponer que es numerico) p-value<alfa. Rechazamos Ho, ph.ecog aumenta el riesgo de morir de cancer en 85% respecto a la categoria anterior.

## Cargar una nueva base de datos

```python
from lifelines.datasets import load_rossi
```

```python
rossi = load_rossi()
```

```python
rossi.head()
```

```python
cph2 = CoxPHFitter().fit(rossi, 'week', 'arrest')
cph2.print_summary()
```

**fin**, p-valor: 0.05, no es menor a alfa, no rechazamos Ho, no tiene efecto darle dinero cuando lo liberas sobre el riesgo de volver a ser arrestado

**age**, p-value: 0.01< alfa, rechazamos Ho, la edad reduce el riesgo de volver a ser arrestado. Por cada anio de edad que se cumple, se reduce el riesgo en 6%.

**race**, p-value:0.31>alfa, no rechazamos Ho, ser negro no tiene efecto sobre el riesgo de volver a ser arrestado.

**wexp, mar, paro**, p-value>alfa, no rechazamos Ho, no tienen efecto en el riesgo de volver a ser arrestado. 

**prio**, p-value: 0.005<alfa, rechazamos Ho, por cada vez que te han arrestado el riesgo de que te vuelvan a arrestar aumenta 10%.

## IMPORTANTE

Signo coef -> aumenta/disminuye riesgo

exp(coef) -> tamanio del efecto

exp(coef)>1 -> exp(coef)-1

exp(coef)<1 -> 1-exp(coef)
