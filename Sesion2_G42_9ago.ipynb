## NUMPY

Numpy sirve para manejar numeros en Python ya sea matrices, vectores, etc. 

Siempre se importara de la siguiente forma:

import numpy as np

```python
import numpy as np
```

Si da error hay que instalar numpy

La estructura básica de Numpy son los arreglos (arrays), que son una colección de objetos del mismo tipo (numericos)

```python
#crear una lista con la venta de la semana de paletas de helado 
ventas = [12, 23, 34, 45, 56]
```

```python
#convertir lista en arreglo de numpy
#array convierte una lista en un arreglo
ventas = np.array(ventas)
#Guardo en ventas para actualizar su contenido
```

```python
type(ventas)
```

```python
#ventas ya es un arreglo de numpy, ya le puedo asignar las funciones de numpy
#solo puedo aplicar las funciones de numpy en arreglos

#Por ejemplo, la media

np.mean(ventas)
```

```python
#Suma

np.sum(ventas)
```

```python
#Ventas mas grandes

np.max(ventas)
```

```python
#Valor mínimo

np.min(ventas)
```

```python
#Python usa una formula de varianza con ddof si ddof= 0 varianza normal y si es igual a 1 es muestral

#Varianza poblacional 
#Si no le pongo ddof python supone que es cero
np.var(ventas)
```

```python
#Varianza muestral
np.var(ventas, ddof=1)
```

```python
#Desviación estándar poblacional
np.std(ventas)
```

```python
#Desviación estándar muestral
np.std(ventas, ddof=1)
```
