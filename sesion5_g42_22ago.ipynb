```python
## Importar librerias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# datos

personas = ['Jorge', 'Raúl', 'Ana', 'Lissa']
edades = [27, 16, 21, 30]

# Crear un diccionario 

dic1 = {'Municipio':['Escobedo', 'Apodaca', 'Guadalupe', 'Santiago'], 'Nacimientos': [123, 345, 278, 321], 'Muertes': [23, 17, 56, 32], 'Matrimonios': [76, 45, 89, 12]}
datos_mun = pd.DataFrame(dic1)
```

```python
datos_mun
```

```python
personas, edades
```

```python
# Gráfica

plt.bar(personas, edades, color = ['lightblue', 'lightpink', 'orange', 'gray'])
plt.title('Edades')
plt.xlabel('Nombres')
plt.ylabel('Edad en años')
plt.show()
```

```python
# Para hacer una gráfica con los municipios y los matrimonios

plt.bar(datos_mun['Municipio'], datos_mun['Matrimonios'])
plt.title('Matrimonios en N.L en 2000')
plt.xlabel('Municipios')
plt.ylabel('Numero de matrimonios')
plt.show()
```

```python
## Hacer 2 graficas juntas
# 1. Darle nombre al pizarron 
# 2. darle nombre a las graficas
# plt.subplots = Como lo queremos acomodar

# fig es el nombre del pizarron

fig, (graf1, graf2)= plt.subplots(1,2)

#Dos graficas acomodadas en dos filas y 1 columna

## GRAFICA 1

graf1.bar(personas, edades)

## GRAFICA 2

graf2.bar(datos_mun['Municipio'], datos_mun['Matrimonios'])

plt.show()
```

```python
#Tenemos ya el dataframe datos_mun
#Pandas hace una barra para cada columna
#Pandas por default pone en el eje X al index

datos_mun.plot.bar();
```

```python
# Si solo quiero la columna de matrimonios

datos_mun['Matrimonios'].plot.bar();
#personalizar
plt.title('Matrimonios en N.L')
plt.xlabel('Municipios')
plt.ylabel('Cantidad de matrimonios')
plt.show()
```

```python
# Hacer municipios index

datos_mun = datos_mun.set_index('Municipio')
datos_mun['Matrimonios'].plot.bar();
#personalizar
plt.title('Matrimonios en N.L')
plt.xlabel('Municipios')
plt.ylabel('Cantidad de matrimonios')
plt.show()
```

```python
## Grafica de barras apiladas

datos_mun.plot.bar(stacked=True);
```

```python
# Grafico de lineas

datos_mun.plot.line();
```

```python
# Grafica de areas

datos_mun.plot.area();
```
